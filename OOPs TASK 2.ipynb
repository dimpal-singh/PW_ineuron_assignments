#Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Default value for odometer reading

    def get_description(self):
        return f"{self.year} {self.make} {self.model}"

    def read_odometer(self):
        return f"This car has {self.odometer_reading} miles on it."

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles):
        self.odometer_reading += miles

# Creating objects (instances) of the Car class
my_car = Car("Toyota", "Corolla", 2022)
friend_car = Car("Honda", "Civic", 2020)

# Accessing attributes and methods of objects
print(my_car.get_description())  # Output: 2022 Toyota Corolla
print(friend_car.get_description())  # Output: 2020 Honda Civic

my_car.update_odometer(1000)
print(my_car.read_odometer())  # Output: This car has 1000 miles on it.

friend_car.increment_odometer(500)
print(friend_car.read_odometer())  # Output: This car has 500 miles on it.
```

#Q2. Name the four pillars of OOPs.

1.Encapsulation: Encapsulation is the bundling of data (attributes) and methods (functions) that operate on that data into a single unit called a class. It helps in hiding the internal state of an object and only exposes the necessary information and functionalities through public interfaces.

2.Inheritance: Inheritance is the mechanism by which one class (subclass or derived class) inherits properties and behaviors (attributes and methods) from another class (superclass or base class). It promotes code reusability and establishes a hierarchy among classes.

3.Polymorphism: Polymorphism allows objects to take on multiple forms or have multiple behaviors based on the context. It enables a single interface (method) to be used for different data types or objects, providing flexibility and extensibility in programming.

4.Abstraction: Abstraction is the process of hiding complex implementation details and exposing only the essential features of an object. It focuses on what an object does rather than how it does it, making the system easier to understand and maintain. Abstraction is often achieved through abstract classes and interfaces.

#Q3. Explain why the __init__() function is used. Give a suitable example.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"My name is {self.name} and I am {self.age} years old."

# Creating instances of the Person class
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

# Accessing attributes and methods of objects
print(person1.introduce())  # Output: My name is Alice and I am 30 years old.
print(person2.introduce())  # Output: My name is Bob and I am 25 years old.
```

#Q4. Why self is used in OOPs?

```python
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        return f"This car is a {self.make} {self.model}."

# Creating instances of the Car class
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

# Accessing attributes and calling methods using self
print(car1.display_info())  # Output: This car is a Toyota Corolla.
print(car2.display_info())  # Output: This car is a Honda Civic.
```

#Q5. What is inheritance? Give an example for each type of inheritance.

```python
class Animal:
    def speak(self):
        return "The animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        return "The dog barks"

# Example usage
dog = Dog()
print(dog.speak())  # Output: The animal makes a sound
print(dog.bark())   # Output: The dog barks
```

```python
class A:
    def method_a(self):
        return "Method A from class A"

class B:
    def method_b(self):
        return "Method B from class B"

class C(A, B):  # C inherits from A and B
    def method_c(self):
        return "Method C from class C"

# Example usage
obj_c = C()
print(obj_c.method_a())  # Output: Method A from class A
print(obj_c.method_b())  # Output: Method B from class B
print(obj_c.method_c())  # Output: Method C from class C
```

```python
class A:
    def method_a(self):
        return "Method A from class A"

class B(A):  # B inherits from A
    def method_b(self):
        return "Method B from class B"

class C(B):  # C inherits from B
    def method_c(self):
        return "Method C from class C"

# Example usage
obj_c = C()
print(obj_c.method_a())  # Output: Method A from class A
print(obj_c.method_b())  # Output: Method B from class B
print(obj_c.method_c())  # Output: Method C from class C
```

```python
class Animal:
    def speak(self):
        return "The animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        return "The dog barks"

class Cat(Animal):  # Cat inherits from Animal
    def meow(self):
        return "The cat meows"

# Example usage
dog = Dog()
cat = Cat()
print(dog.speak())  # Output: The animal makes a sound
print(dog.bark())   # Output: The dog barks
print(cat.speak())  # Output: The animal makes a sound
print(cat.meow())   # Output: The cat meows
```
